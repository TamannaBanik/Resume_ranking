```python
import json
import pandas as pd
from tqdm import tqdm
from collections import Counter
```

```python
candidate_path = "/kaggle/input/datasets/tamannabanik/candidates-jsonl//candidates.jsonl"
candidates = []
with open(candidate_path, "r") as f:
    for line in tqdm(f):
        candidates.append(json.loads(line))
print("Total Candidates:",len(candidates))
```

```python
skill_counter = Counter()
for candidate in candidates:
    scores = candidate.get("redrob_signals",{}).get("skill_assessment_scores",{})
    for skill in scores:
        skill_counter[skill] += 1
```

```python
print("Unique Skills:",len(skill_counter))
```

```python
skills_df = pd.DataFrame(skill_counter.items(),columns=["Skill","Count"])
skills_df = skills_df.sort_values("Count",ascending=False)
skills_df            
```
